```python
import tensorflow as tf
import numpy as np
import pandas as pd
import os
import skimage.data
import matplotlib.pyplot as plt
from skimage import transform
from skimage.color import rgb2gray

import warnings
warnings.filterwarnings("ignore")
```

```python
def get_abs_paths(path):
    for file_name in os.listdir(path):
        yield os.path.join(path, file_name)

def load_data(data_directory):
    """
    data_directory is a path till training and testing it is the absolute path,
    where there are folders named 00000, 00001 and we have ppm files in it they are images.
    so we will create two lists images(x) and labels(y)
    """
    images = []
    labels = []
    
    for dir in get_abs_paths(data_directory):
        if os.path.isdir(dir):
            for image in get_abs_paths(dir):
                if image.endswith('ppm'):
                    images.append(skimage.data.imread(image))
                    labels.append(int(dir.split("/")[-1]))
    return images,labels
```

```python
root_path = "/home/SONY/Coursera_ML_Python/tensorflow/Belgian"
training_path = os.path.join(root_path, "Training")
testing_path = os.path.join(root_path, "Testing")

images, labels = load_data(training_path)
images = np.array(images)
labels = np.array(labels)


print(images.size)
print(images.size)
print(labels.shape)

print(images[0].shape)
print(images[1].shape)
print(images[2].shape) #Don't worry if the images are not of same size,they are suppossed to be this way
```

```python
#Let's transform the images to 28*28 Some warnings will shootup, ignore them
images28 = [transform.resize(image, (28,28)) for image in images]
```

```python
#grey scale conversion, we will check what happens if we don't do that. For now let's do it

images28 = np.array(images28)
images28 = rgb2gray(images28)

random_images = [39, 546, 890, 2356, 3200, 4000]

for i in range(len(random_images)):
    plt.subplot(2,3, i+1)
    plt.axis("off")
    plt.imshow(images28[random_images[i]], cmap="gray") #cmap is important,otherwise some heated maps will be shown like purple

plt.show()
```

```python
#Now tf will begin, remember this is not CNN, it is simple NN
"""
Remember the structure , that is more important than function names, structure can't be googled functions can be
1) Define place holders for x,y
2) If needed flatten the image
3) define logits (fully connected layer), (This may come later if network is deep or CNN)
4) Define loss function
5) Define an optimizer (Usually Adam)
6) Convert logits to labels, make the max element as one
"""
#This are just placeholders, kind of skeleton, the values will be filled in when we put them in the tf.session
#So, just define the structure here, do not connect them to images28 or labels
x = tf.placeholder(dtype= tf.float32, shape=[None,28,28])
y = tf.placeholder(dtype= tf.int32, shape=[None]) #Single dimension 1d array

images_flat = tf.contrib.layers.flatten(x) # [None,28,28] to [None, 784]

#Fully Connected layer
logits = tf.contrib.layers.fully_connected(images_flat, 62, tf.nn.relu) #Self explainatory

#Loss function
loss = tf.reduce_mean(tf.nn.sparse_softmax_cross_entropy_with_logits(logits=logits, labels=y)) #Loss between logit and label

#Optimizer, Adam optimizer with learning rate and what to minimize
train_op = tf.train.AdamOptimizer(learning_rate=0.001).minimize(loss)

#Convert logits to label indexes
correct_pred = tf.argmax(logits, 1)

#Define accuracy, Mean as accuracy measure
accuracy = tf.reduce_mean(tf.cast(correct_pred, tf.float32))

print("images_flat: ", images_flat)
print("logits: ", logits)
print("loss: ", loss)
print("predicted_labels: ", correct_pred) #All correct till here, we have defined our model, then we will train it
```

```python
tf.set_random_seed(1234)

sess = tf.Session()
sess.run(tf.global_variables_initializer())
for i in range(2000): #1000 epoches
    _, loss_value = sess.run([train_op, loss], feed_dict={x: images28, y:labels}) #Put values in only placeholders
                                                                                  #Other values are calculated base on these 2
    if i%50 == 0:
        print("loss ",loss_value) #Loss gets reduced after epochs
```

```python
#Let's evaluate the model
test_images, test_labels = load_data(testing_path)
test_images = np.array(test_images)
test_labels = np.array(test_labels)

test_images28 = [transform.resize(image, (28,28)) for image in test_images]
test_images28 = rgb2gray(np.array(test_images28))

print(test_images28[0].shape)
```

```python
predicted = sess.run([correct_pred], feed_dict={x: test_images28})[0]#This one is the logits returned,so we areusing this,no questions asked

#calculate the correct matches
match_count = sum([int(y == y_) for y, y_ in zip(test_labels, predicted)])

# Calculate the accuracy
accuracy = match_count / len(test_labels)

print(accuracy) #67% in 300 matches it correctly, Nice Job 60% in 2000 , are we overfitting?
                #You are seeing values decreasing in the loss, but it is overfitting it on training_data and
                #Fails miserably on the test data.
sess.close()
```
